```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset : https://archive.ics.uci.edu/dataset/236/seeds

```python
cols = ["area", "perimeter", "compactness", "length", "width", "asymmetry", "groove", "class"]
df = pd.read_csv("seeds_dataset.txt", sep=r"\s+", names=cols)
df.head()
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
x = "perimeter"
y= "asymmetry"
X = df[[x,y]].values
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
```

```python
clusters = kmeans.labels_
```

```python
clusters
```

```python
df["class"].values
```

```python
clusters_df = pd.DataFrame(
    np.hstack((X, clusters.reshape(-1, 1))),
    columns=[x,y, "class"]
)
```

```python
sns.scatterplot(x=x, y=y, hue="class", data=clusters_df)

plt.plot()
```
